## 1.6. Series 의 연산

Series 를 계산에 사용할 때엔 재미있는 특징이 있습니다.

바로 broadcasting 인데요.

방송을 송출하는 듯한 이 것은 무엇인지 예를 통해 봅시다.

```python
a = pd.Series([1, 2, 3])

a + 1
```

분명 a 는 1, 2, 3이 들어있는 Series 이기에,  (1, 2, 3) + 1 이라는 이상한 구조처럼 보이지만, 

계산되는 대상이 scalar 인 경우, 이를 자동적으로 Series 에 맞춰줍니다.

그러니까 실제로는, (1, 2, 3) + (1, 1, 1) 이 되어서 결과인 (2, 3, 4) 가 나온 것이죠.

이런 계산은 사칙연산 ( +, -, *, /, //, %) 뿐만 이나리 논리 비교( <, >, >=, <=, !) 에서도 동일하게 작용합니다.

하나 더 만들어볼게요.

```python
a = pd.Series(np.arange(0, 10))
# a = [0, 1, 2, ..., 8, 9] % [2, 2, 2, 2, ... 2]
print(a % 2 == 0)
```

위의 코드는 0부터 9까지의 수 중 2로 나눈 나머지가 0인, 즉 짝수인 경우를 확인하는 코드입니다.

a 는 전체 Series 를 의미하지만, 이를 식으로 표현할 때엔 마치 하나의 수인 것처럼 다룰 수 있죠.

이런 연산을 vector 화된 연산 (vectorization )이라고 하는데, for 문을 사용하는 것보다 훨씬 효율적인 접근입니다.

나중에 또 다룰 기회가 있을 겁니다.

---

동일한 계산을 함수를 사용해서도 가능합니다.

예를 들어 다음과 같죠.


```python
fb_likes.add(1) # fb_likes + 1
fb_likes.sub(2) # fb_likes - 2
```

첫 번째는 fb_likes + 1, 두 번째는 fb_likes - 2와 같습니다.

그러면 둘 사이엔 어떤 차이가 있는 것일까요?

함수를 이용해서 계산할 때엔 fill_value 라는 인자를 넘겨줄 수 있는데, 이를 통해 na 값을 처리할 수 있습니다.

na 값은 어떤 사칙연산을 적용하더라도 na 가 나옵니다. 아래의 결과를 보죠.


```python
np.nan - 1
```

그런데 fill_value 를 이용하면, fill_na 의 효과가 나타나면서 계산이 가능합니다.

다음과 같이 말이죠.


```python
fb_likes.sub(2, fill_value=0)[fb_likes.isna()]
```

위의 내용은 na 값을 가진 것에 0을 채워 넣은 뒤 sub(2), 즉 2를 뺀 것이 됩니다.

여기서 기억할 두 가지

1. na 는 계산 불가능하다.
2. vector 연산에서 na 가 무시되는 것을 막기 위해서는 fillna 를 이용해 값을 먼저 채우거나, 계산 함수에서 fill_value 를 써준다.

## 1.7. 끝나기 전의 팁 NA는 비교불가

NA 값은 NA 끼리도 비교가 불가능합니다.

아래의 출력을 보죠.

```python
print(np.nan != np.nan)
print(np.nan > 1)
print(np.nan < 1)
print(np.nan == 1)
```

오직 같지 않다! 만 참이 나오고 나머지가 다 거짓이 되죠.

그러니까, S나 DF를 비교할 때도 null 값을 처리하지 않으면 문제가 됩니다.

아래와 같이 말이죠.

```python
import pandas as pd
a = pd.Series([1, 2, 3])
b = pd.Series([1, 2, 3])
c = pd.Series([1, np.nan, 3])
d = pd.Series([1, np.nan, 3])
```

```python
#무사통과하는 a == b
print(a == b)
print()
# 1번 인덱스를 잘 보세요.
print(c == d)
```

그래서 null 값이 섞여 있는 객체를 비교할 땐, __equals__ 메서드를 사용해야만 합니다.

```python
# equals 는 두 객체가 동일한 요소들을 갖고 있는지 확인하는 메서드입니다.
print(a.equals(b))
print(c.equals(d))
```
